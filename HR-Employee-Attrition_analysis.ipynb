```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
le=LabelEncoder()
df['Attrition']=le.fit_transform(df['Attrition'])
x=dfn
y=df['Attrition'].loc[dfn.index]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier
lo=LogisticRegression()
knn=KNeighborsClassifier()
svc=SVC()
gb=GaussianNB()
ml=MultinomialNB()
dtc=DecisionTreeClassifier()
l=[lo,knn,dtc,svc,gb,ml]
for i in l:
    i.fit(x_train, y_train)
    pred = i.predict(x_train)
    acc = np.round(accuracy_score(y_train, pred) * 100, 2)
    print(f"====={ i } ==================================== Accuracy_Score:{ np.round(accuracy_score(y_train, pred) * 100, 2)}%")
    report = classification_report(y_train, pred, output_dict=True)
    #rint(#pd.DataFrame(report).transpose())
    print(classification_report(y_train, pred))
    print("================================================================================")
    print('\n')
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(knn,x,y,cv=9)
print('cv score of SVC ',np.sort(score))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier()
x = {'n_estimators':[50,100], 'max_depth':[5,10]}

grid = GridSearchCV(estimator=rf, param_grid=x, cv=3)
grid.fit(x_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': [5, 10, 15, None]
}

random_search = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=5, cv=3, scoring='accuracy', random_state=42)
random_search.fit(x_train, y_train)

print("Best Params:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```
